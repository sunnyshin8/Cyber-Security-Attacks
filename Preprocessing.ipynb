```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the dataset
df = pd.read_csv('data/cybersecurity_attacks.csv')
```

```python
# checking the shape of the dataset
print(df.shape)
```

```python
df.head()
```

```python
# Finding the missing values
df.isnull().sum()
```

```python
# Replacing the missing values with 'None'
df['Malware Indicators'].fillna('None', inplace=True)
df['Alerts/Warnings'].fillna('None', inplace=True)
df['Firewall Logs'].fillna('None', inplace=True)
df['IDS/IPS Alerts'].fillna('None', inplace=True)

# Replacing the missing values with 'No Proxy'
df['Proxy Information'].fillna('No Proxy', inplace=True)
```

```python
df.isnull().sum()
```

```python

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
# Using Timestamp to create new features
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month_name()
df['Date'] = df['Timestamp'].dt.day
df['Hour'] = df['Timestamp'].dt.hour
df['Day'] = df['Timestamp'].dt.day_name()
```

```python
df.head()
```

```python
df.to_csv('data/cybersecurity_attacks_cleaned.csv', index=False)
```
